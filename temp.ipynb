```python
from bp import FactorGraph
import networkx as nx
import matplotlib.pyplot as plt 
```

```python
self = FactorGraph()

self.add_variable_node('a', 2)
self.add_variable_node('b', 2)
self.add_variable_node('c', 2)
self.add_variable_node('d', 2)

self.add_factor_node('f1', ['a', 'b'], [[0.1, 0.9], [0.9, 0.1]])
self.add_factor_node('f2', ['b', 'c', 'd'], [[[0.1, 0.9], [0.9, 0.1]], [[0.1, 0.9], [0.9, 0.1]]])
self.add_factor_node('f3', ['c'], [0.1, 0.9])

pos = {'a': (0, 0), 
    'f1': (1, 0), 
    'b': (2, 0), 
    'f2': (3, 0), 
    'c': (4, 1), 
    'd': (4, -1), 
    'f3': (5, 1)}

inverted_pos = {key: (value[1], value[0]) for key, value in pos.items()}
```

```python
plt.figure(figsize=(10, 5))
nx.draw_networkx(self.g, pos, nodelist=self.names_var_nodes, node_shape='o')
nx.draw_networkx(self.g, pos, nodelist=self.names_fact_nodes, node_shape='s', node_color='grey')
```
